```python
# Reconfigurar para salvar sem falhas
import pandas as pd
from geopy.geocoders import Nominatim
import time
```

```python
# Recarregar o arquivo
file_path = r'C:\Users\Administrator\Documents\projetos_py\unlimitail\Blocos e Trios elétricos - CEP.xlsx'
df = pd.read_excel(file_path)

# Configurar geolocalizador
geolocator = Nominatim(user_agent="geoapi")

df.head()
```

```python

# Função para obter coordenadas
def obter_coordenadas(cep):
    try:
        location = geolocator.geocode({"postalcode": cep, "country": "Brazil"})
        if location:
            return location.latitude, location.longitude
    except Exception as e:
        print(f"Erro ao buscar coordenadas para o CEP {cep}: {e}")
    return None, None

# Adicionar colunas de latitude e longitude
df['Latitude'], df['Longitude'] = zip(*df['CEP'].astype(str).apply(obter_coordenadas))

df.head()
```

```python
df_na_values = df[df[['Latitude', 'Longitude']].isna().any(axis=1)]
df_na_values.head(100)
```

```python
# Salvar o arquivo atualizado
output_path = r'C:\Users\Administrator\Documents\projetos_py\unlimitail\Blocos e Trios elétricos - CEP.xlsx'
df.to_excel(output_path, index=False)

# import ace_tools as tools; tools.display_dataframe_to_user(name="Blocos e Trios Elétricos com Coordenadas", dataframe=df)
```
